## __I. Essential Plots in Statistics__

Here are some of the essential plotting methods in statistics, especially in Exploratory Data Analysis (EDA).

```python
# Using the iris dataset
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = load_iris()
x1 = data.data[:, 0] # Sepal Length
x2 = data.data[:, 1] # Sepal Width
y = data.target

print(set(y))
```

### *Scatter Plot*

Scatter plots helps identify the relationship between two variables, highlight outliers, and identify clusters of data.

(Ceb.) Sa pag gamit nato sa *Scatter plot*, makita nato ang relasyon sa duha ka baryable. Bale, makita nato kung naa bay koneksyon base sa porma sa atong datos.

```python
plt.figure(figsize=(12, 8))
plt.scatter(x1, x2, c=y, cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

### *Box Plot*

Helps to describe the distribution or the spread of one or more groups of numeric data.

(Ceb.) Sa pag gamit nato og *Box Plot*, diri nato makita kung unsa ka kalat atong datos.

```python
species = data.target_names
d = []
for target in set(y):
    d.append(x1[y == target])
    
plt.figure(figsize=(12, 6))
plt.boxplot(d, tick_labels=species)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

### *Histograms*

It is a graph showing the number of observations within each given interval. It shows if the data is normal (bell curve) or skewed. It also identifies the spread of data.

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5)) 
for i, target in enumerate(np.unique(y)): 
    # Filter data for the specific species
    feature1_subset = x1[y == target] 
    
    # Plot 2D histogram
    axs[i].hist(feature1_subset, bins=15)
    axs[i].set_title(data.target_names[target])  
    axs[i].set_ylabel(data.feature_names[1])
```

### *Heatmaps*

Heatmaps are used to show the relationship/correlation between two variables.

```python
X = data.data
df = pd.DataFrame(X, columns=[feature for feature in data.feature_names])
corr = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Iris Features correlation matrix')
plt.show()
```

## __II. Statistics Review__

### *Mean*

$$\bar{x}=\frac{\sum x}{n}$$

### *Variance*

__Population Variance__. Uses the entire dataset, and then divides by the total number of data points.
$$\sigma^2=\frac{\sum (x - \mu)}{N}$$

__Sample Variance__. Uses a subset of the dataset, and then divides by the sample size - 1 (Bessel's Correction). This provides a better, unbiased estimate of the true population variance.
$$s^2=\frac{\sum (x - \bar{x})}{N-1}$$

### *Standard Deviation*

__Population Standard Deviation__. 
$$\sigma=\sqrt{\frac{\sum (x - \mu)}{N}}$$

__Sample Standard Deviation__. 
$$s=\sqrt{\frac{\sum (x - \bar{x})}{N-1}}$$
