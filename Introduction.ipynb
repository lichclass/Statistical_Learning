## I. History of Statistical Learning

- (1936) Introduction of *Linear Discriminant Analysis*
- (1940s) Proposal of *Logistic Regression*
- (1970s) Developed the *Generalized Linear Model* to describe an entire class of statistical learning methods
- (mid-1980s) *Classification* and *Regression Trees* were developed, followed by *Generalized Additive Models*
- (1980s) Popularized *Neural Networks* and *Support Vector Machines (SVM)*
